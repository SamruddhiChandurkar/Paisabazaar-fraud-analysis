## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Import Libraries
import pandas as pd
import numpy as np

# Load Dataset
df = pd.read_csv("dataset.csv")

# 1. Initial Overview

print(df.shape)
print(df.info())
print(df.head())

# 2. Remove Unnecessary Columns

df.drop(['ID','Customer_ID','Name','SSN'], axis=1, inplace=True, errors='ignore')

# 3. Check Missing Values

print(df.isnull().sum())


# 4. Handle Missing Values

# Option 1: Fill missing values forward
df.fillna(method='ffill', inplace=True)

# OR Option 2 (if needed):
# df.dropna(inplace=True)

# 5. Fix Data Types
cols_to_convert = ['Age','Annual_Income','Outstanding_Debt',
                   'Num_of_Loan','Num_of_Delayed_Payment']

for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 6. Handle Incorrect / Negative Values

df = df[df['Age'] > 0]


# 7. Remove Duplicates

print("Duplicates:", df.duplicated().sum())
df.drop_duplicates(inplace=True)

# 8. Clean Categorical Columns
df['Occupation'] = df['Occupation'].str.strip()
df['Credit_Mix'] = df['Credit_Mix'].str.strip()
df['Payment_Behaviour'] = df['Payment_Behaviour'].str.strip()

# 9. Check Unique Values

for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")


# 10. Final Dataset Check

print("Final Shape:", df.shape)
print(df.describe())
````

### What all manipulations have you done and insights you found?


Several data preprocessing and cleaning steps were performed to prepare the dataset for Exploratory Data Analysis:

1. Removed unnecessary columns such as ID, Customer_ID, Name, and SSN because they do not contribute to the analysis.

2. Checked unique values in each column to understand the diversity of data.

3. Handled missing values by identifying null values and removing or filling them where necessary.

4. Converted incorrect data types (such as Age, Annual_Income, and Outstanding_Debt) into numeric format to ensure accurate analysis.

5. Removed duplicate records to avoid biased results.

6. Performed statistical summary analysis to understand the distribution, mean, and variation of numerical variables.

7. Created visualizations such as histograms, boxplots, scatterplots, and heatmaps to explore relationships between variables.

These manipulations helped in cleaning the dataset and making it suitable for detailed exploratory analysis.
